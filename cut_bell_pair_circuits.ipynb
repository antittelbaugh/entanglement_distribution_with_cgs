# Cut Bell pair circuits

This notebook shows the circuits that were used to create the cut Bell pairs. These circuits are stored in QPY format, the Qiskit serializer.

```python
from qiskit import qpy
```

## One cut Bell pair

Here we show the circuit for a single cut Bell pair. This QPD has $\gamma=3$.

```python
with open("data/one_bell_pair.qpy", "rb") as fd:
    circuit1 = qpy.load(fd)[0]

circuit1.draw("mpl", fold=False)
```

The parameter sets are in the file below.

```python
with open("data/one_qpd_bell_pair_param_values.txt") as fin:
    params1 = [[float(val) for val in line.replace("\n","").split(" ")] for line in fin.readlines()]

for params in params1:
    print(params)
```

These parameters can be assigned in the standard way. The weights of the QPD are given in the paper.

```python
circuit1.assign_parameters(params1[0], inplace=False).draw("mpl")
```
